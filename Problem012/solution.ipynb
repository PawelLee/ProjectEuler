# Highly divisible triangular number

The sequence of triangle numbers is generated by adding the natural numbers.

What is the value of the first triangle number to have over five hundred divisors?

```python
def tri_number(input_int):
    return(input_int*(input_int+1)//2)
```

```python
import math
def prime_factorisation(n):
    factor_list = []
    while not n%2:
        n //= 2
        factor_list.append(2)
    prime = False
    while not prime:
        prime = True
        if math.ceil(math.sqrt(n)) + 1 <= 3:
            break        
        for i in range(3, math.ceil(math.sqrt(n))+1, 2):
            if not n%i:
                n //= i
#                 print(factor_list,'appending', i)
                factor_list.append(i)
                prime = False
                break
    factor_list.append(n)

    return factor_list
```

```python
import collections

def number_of_divisors(n):
    prime_factors = prime_factorisation(n)
    factors_multiplicity = collections.Counter(prime_factors)
    divisor_count = 1
    for exponent in factors_multiplicity.values():
        divisor_count *= (exponent + 1)
    return(divisor_count)
```

```python
i = 3
while True:
    i+=1
#     if not i%10000:
#         print(i, tri_number(i), number_of_divisors(tri_number(i)))
    if number_of_divisors(tri_number(i)) > 500:
        print("solution: ", tri_number(i))
        break
```
